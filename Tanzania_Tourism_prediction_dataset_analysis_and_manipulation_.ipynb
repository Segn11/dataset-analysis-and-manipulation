```python
df['total_male'] = df['total_male'].fillna(0)
```

```python
# Strip whitespace in string columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()
```

```python
# Convert numeric-like columns
numeric_cols = [ 'total_male', 'night_mainland', 'night_zanzibar', 'total_cost']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Feature engineering
# df['total_people'] = df['total_female'] + df['total_male']
df['total_nights'] = df['night_mainland'] + df['night_zanzibar']
# df['cost_per_person'] = df['total_cost'] / df['total_people']
df['cost_per_night'] = df['total_cost'] / df['total_nights']
```
